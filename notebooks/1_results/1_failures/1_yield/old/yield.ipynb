# Yield

* per site in cortex
* 10 minutes of recording

method: We divide the unit yield by the number of channels because different experiments had a different number of sites in the cortex, while some sites were in the white matter or outside the brain.

The unit yield is a first metrics to quantify the dark matter problem. It reveals the the extreme subsampling of the population within 50 microns of the probe
