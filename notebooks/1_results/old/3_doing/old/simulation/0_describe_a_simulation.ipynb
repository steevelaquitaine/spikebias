* `cNAC`: classical non-accomodating cells
* `bSTUT`: bursting stuttering
* `bNAC`: bursting non-accomodating cells

* `cACint`: classical accomodating interneuron?

* `cADpyr`: classical accomodating delayed pyramidal?
