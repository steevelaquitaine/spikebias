```python
# plot
fig, ax =  plt.subplots(1,1, figsize=(20,20))
Npx = MEA.return_mea("Neuropixels-384")
MEA.plot_probe(ax=ax, mea=Npx);
```

```python
# check vertical distance between rows of electrodes (only vertical displacement as reported in the literature) 
y_pitch = np.sqrt(sum((Npx.positions[1] - Npx.positions[0])**2))
print("y_pitch:", y_pitch)

# check horizontal distance between columns of electrodes (only horizontal displacement as reported in the literature) 
contact0 = Npx.positions[0]
contact96 = np.array([0, Npx.positions[96,1], Npx.positions[0][2]])
x_pitch = np.sqrt(sum((contact96 - contact0)**2))
print("x_pitch:", x_pitch)

# check center-to-center (euclidean) distance between nearest electrodes (as reported in the literature) 
contact0 = Npx.positions[0]
contact96 = Npx.positions[96]
euclidean = np.sqrt(sum((contact96 - contact0)**2))
print("center-to-center:", euclidean)
```
