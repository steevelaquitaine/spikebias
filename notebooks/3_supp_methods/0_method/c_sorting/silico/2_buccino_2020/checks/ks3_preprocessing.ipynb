# Check KS 3.0 preprocessed traces

purpose: check that spikes are visible.

### Setup

```python
import spikeinterface as si
from matplotlib import pyplot as plt

# path where Kilosort 3.0 saves preprocessed file when sorting is run
ks3_preprocessed_path = "/gpfs/bbp.cscs.ch/project/proj85/scratch/laquitai/2_bernstein_2023/sorting/0_silico/buccino/KS3_output/temp_wh.dat"
```

### Plot

```python
# load recording
Recording = si.read_binary(
    ks3_preprocessed_path, sampling_frequency=40000, num_chan=384, dtype="int16"
)
traces = Recording.get_traces()

# plot one second segment
plt.plot(traces[: 2 * 30000, 0])
plt.show()
plt.plot(traces[: 2 * 30000, 1])
plt.show()
plt.plot(traces[: 2 * 30000, 2])
plt.show()
plt.plot(traces[: 2 * 30000, 3])
plt.show()
plt.plot(traces[: 2 * 30000, 4])
plt.show()
plt.plot(traces[: 2 * 30000, 5])
plt.show()
plt.plot(traces[: 2 * 30000, 6])
plt.show()
plt.plot(traces[: 2 * 30000, 7])
```

```python
# zoom on a spike
plt.plot(traces[:3000, 0])
plt.show()
plt.plot(traces[:3000, 1])
plt.show()
plt.plot(traces[:3000, 2])
plt.show()
plt.plot(traces[:3000, 3])
plt.show()
plt.plot(traces[:3000, 4])
plt.show()
plt.plot(traces[:3000, 5])
plt.show()
plt.plot(traces[:3000, 6])
plt.show()
plt.plot(traces[:3000, 7])
plt.show()
plt.plot(traces[:3000, 180])
plt.show()
plt.plot(traces[:3000, 380])
```
