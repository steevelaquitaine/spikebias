```python
import scipy
import numpy as np

np.random.seed(0)

n_sites = 200

# 200 sites and 100 frequencies
f1 = np.random.rand(200, 100000)
f2 = np.random.rand(200, 100000)

# calculate absolute area numerical integration
# for all site pairs
areas = []
for v_i in range(n_sites):
    for s_i in range(n_sites):
        areas.append(sum(np.absolute(f1[v_i, :] - f2[s_i, :])))
areas = np.array(areas)

# unit-test
assert len(areas) == 200 * 200, "pair count is wrong"

# two-sided one sample t-test
# This is a test for the null hypothesis that the expected value (mean)
# of a sample of independent observations a is equal to the 0.
result = scipy.stats.ttest_1samp(areas, 0, alternative="two-sided")
result
```

```python
from matplotlib import pyplot as plt

plt.hist(areas, bins=20)
```
