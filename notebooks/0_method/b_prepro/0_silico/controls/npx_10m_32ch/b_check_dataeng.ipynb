**In brief**: 
* A simulation (configured in a single `BlueConfig`) is chunked into 30 pieces each treated by a different computer node with a rank id, such that each piece can fit into RAM (when `bluepy.Simulation` is called) to extract trace and spike data.
* For multiple simulations (a campaign of several `BlueConfig`), this is done in parallel.
* For a single simulation, it can be done sequentially by running `chunking.py` module
